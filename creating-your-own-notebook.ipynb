```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
my_filepath="../input/startups/50_Startups.csv"
```

```python
startup_data=pd.read_csv(my_filepath)
```

```python
startup_data.head()
```

```python
plt.figure(figsize=(16,6))
plt.title("Startup Companies")
sns.lineplot(data=startup_data)
```

```python
plt.figure(figsize=(16,6))
plt.title("R&D Spend Vs Profit")
sns.barplot(x=startup_data['R&D Spend'],y=startup_data['Profit'])
```

```python
plt.figure(figsize=(16,6))
plt.title("Marketing Spend Vs Profit")
sns.barplot(x=startup_data['Marketing Spend'],y=startup_data['Profit'])
```

```python
plt.figure(figsize=(16,6))
plt.title("Correlation between R&D spend and Profit and belonging from which state")
sns.scatterplot(x=startup_data['R&D Spend'],y=startup_data['Profit'],hue=startup_data['State'])
```

```python
plt.figure(figsize=(16,6))
plt.title("Correlation between Marketing spend and Profit and belonging from which state")
sns.scatterplot(x=startup_data['Marketing Spend'],y=startup_data['Profit'],hue=startup_data['State'])
```

```python
plt.figure(figsize=(16,6))
plt.title("distribution of Profit")
sns.kdeplot(data=startup_data['Profit'],shade=True)
```

```python
sns.jointplot(x=startup_data['R&D Spend'],y=startup_data['Marketing Spend'],kind='kde')
```

```python
startup_data['State'].unique()
```

```python
data_1=startup_data[startup_data['State']=='New York']
data_2=startup_data[startup_data['State']=='California']
data_3=startup_data[startup_data['State']=='Florida']
```

```python
sns.distplot(a=data_1['Profit'],label='New York',kde=False)
sns.distplot(a=data_2['Profit'],label='California',kde=False)
sns.distplot(a=data_3['Profit'],label='Florida',kde=False)
plt.title("Histogram of Profit by States")
plt.legend()
```

```python
sns.distplot(a=data_1['Profit'],label='New York')
sns.distplot(a=data_2['Profit'],label='California')
sns.distplot(a=data_3['Profit'],label='Florida')
plt.title("Histogram of Profit by States with kde")
plt.legend()
```
